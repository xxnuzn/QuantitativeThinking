```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 매출 데이터 예시
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [1000, 1500, 2000, 2500, 3000, 3500]
}

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 월별 매출 그래프 그리기
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='b')
plt.title('월별 매출')
plt.xlabel('월')
plt.ylabel('매출 (단위: 천원)')
plt.grid(True)
plt.show()

# 간단한 통계 분석
mean_sales = df['Sales'].mean()  # 평균 매출
median_sales = df['Sales'].median()  # 중앙값
std_sales = df['Sales'].std()  # 표준편차

print(f"평균 매출: {mean_sales:.2f} 천원")
print(f"중앙값 매출: {median_sales:.2f} 천원")
print(f"표준편차: {std_sales:.2f}")
```
